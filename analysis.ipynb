```python
plt.figure(figsize=(8,6))
ax=sns.countplot(data=df2,x="Gender",hue="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution of customers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df2,x="Gender",hue="Gender")

plt.title("Gender Distribution of customers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
ax= sns.countplot(data=df2,x="Gender",hue="Gender")
for i in ax.containers:
     ax.bar_label(i)
plt.title("Gender Distribution of customers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python

plt.figure(figsize=(8,6))
ax=sns.barplot(data=df2,x="Gender",y="Purchase Amount (USD)",estimator=sum , hue="Gender")
for i in ax.containers:
     ax.bar_label(i)
plt.title("Gender Distribution of customers")
plt.xlabel("Gender")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```

```python
df2["Gender"].value_counts().plot(kind='pie', autopct='%.1f%%', explode=(0.1, 0.1), figsize=(6, 6),shadow=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df2,x="Item Purchased",hue="Item Purchased")
plt.title("Age Distribution of customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.boxplot(data=df2 ,x="Item Purchased",y="Purchase Amount (USD)",hue="Category")

plt.title("Age Distribution of customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
